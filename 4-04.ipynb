```python
animal = 'cat'
print("animal:", animal)

def my_func():
    animal = 'dog'
    print("animal in my_func:", animal)

my_func()

print("animal global after my_func:", animal)
```

```python
animal = 'cat'
print("animal:", animal)

def my_func_alter():
    global animal 
    animal = 'dog'
    print("animal in my_func:", animal)

my_func_alter()

print("animal global after my_func:", animal)
```
